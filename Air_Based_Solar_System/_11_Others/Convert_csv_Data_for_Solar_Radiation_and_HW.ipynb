```python
def convert_data_example(csv_name):
    
    def cnv(s):
        try:
            float(s)
        except ValueError:
            return None
        else:
            return float(s)
    
    with open(csv_name, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        next(reader)  # ヘッダーをスキップ
        next(reader)  # ヘッダーをスキップ
        d = [row for row in reader]
    
    # string => float
    cnv_d = []
    for i, items in enumerate(d):
        if len(items) == 1:
            if i < 8760:
                cnv_d += tuple([cnv(item) for item in items])
        else:
            if i < 8760:
                cnv_d += [tuple([cnv(item) for item in items])]
    
    print(cnv_d)
```
