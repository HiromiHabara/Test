平均外気温度 / daily average temperature

```python
def get_theta_ex_Ave(region, day):
    return np.average([EC.get_data(region,day,hour)[0] for hour in range(0, 24, 1)]) 
```

期間平均外気温度 / last 10 days average temperature

```python
def get_theta_ex_prd_Ave(region, day):
    
    nday = EC.get_Annual_Days().index(day) # start the number from 0
    theta_ex_Ave_s = []
    for n in range(nday-10, nday, 1):
        if n < 0:
            theta_ex_Ave_s += [get_theta_ex_Ave(region, EC.get_Annual_Days()[365+n])]
        else:
            theta_ex_Ave_s += [get_theta_ex_Ave(region, EC.get_Annual_Days()[n])]
    
    return np.average(theta_ex_Ave_s)
```

```python
get_theta_ex_prd_Ave('Region6', '1/2')
```

日平均給水温度 / daily average water temperature

```python
def get_theta_wtr(region, day):
    
    a_wtr, b_wtr = {
        'Region1' : (0.6639, 3.466),
        'Region2' : (0.6639, 3.466),
        'Region3' : (0.6054, 4.515),
        'Region4' : (0.6054, 4.515),
        'Region5' : (0.8660, 1.665),
        'Region6' : (0.8516, 2.473),
        'Region7' : (0.9223, 2.097),
        'Region8' : (0.6921, 7.167),
    }[region]
    
    return max(a_wtr * get_theta_ex_prd_Ave(region, day) + b_wtr, 0.5)
```
